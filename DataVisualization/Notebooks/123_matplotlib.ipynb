What if I want to change the ***Line Style*** and ***Line Width*** ?  
The solution is ***linestyle*** and ***linewidth*** parameter.  
  
The allowed values for the linestyle parameter are as follows :  
```python
['solid', 'dashed', 'dashdot', 'dotted']
```

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    linewidth = 3
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    linewidth = 1
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
````

**Problem :** I am not able to see the points.  
**Solution :** Use ***marker*** parameter.

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    marker = 'o'
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    marker = '+'
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
````

**Problem :** I am not sure which line represents ***V Kohli*** and which represents ***RG Sharma***.  
**Solution :** While calling plot for each give labels by the ***label*** parameter and then call the ***legend*** function.

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    marker = 'o',
    label = 'V Kohli'
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    marker = '+',
    label = 'RG Sharma'
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.show()
````

Though ***legend*** function itself identifies the best location for legend to show up, we can customize it using the ***loc*** parameter of the ***legend*** function.

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    marker = 'o',
    label = 'V Kohli'
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    marker = '+',
    label = 'RG Sharma'
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(
    loc = 'upper right'
)
plt.show()
````

What if we want to just see the data in between ***2010*** and ***2014*** ?  
The answer is that we can use the ***xlim*** and ***ylim*** functions to limit the data on ***X-axis*** and ***Y-axis*** respectively.

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    marker = 'o',
    label = 'V Kohli'
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    marker = '+',
    label = 'RG Sharma'
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.xlim(left = 2010, right = 2014)
plt.legend()
plt.show()
````

We can think of ***limiting*** the axes using ***xlim*** and ***ylim*** as cropping the plot.  

What additional things I can do to make the visualiation more ***readable*** and ***intuitive*** ?  
The answer is ***grid*** method, which allows us to show grids on the visualization.

````python
plt.plot(
    sk['index'],
    sk['V Kohli'],
    color = 'gold',
    linestyle = 'dashed',
    marker = 'o',
    label = 'V Kohli'
)
plt.plot(
    sk['index'],
    sk['RG Sharma'],
    color = 'mediumslateblue',
    linestyle = 'dashdot',
    marker = '+',
    label = 'RG Sharma'
)
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
plt.legend()
plt.show()
````

Lets plot some more scatter plots on a very interesting dataset of ***Seaborn*** library, which is ***tips***.

Lets have a look at the dataset structure.

````python
tips_df.head()
````

Lets plot some interesting visualizations using ***tips_df***.  
Below is a scatter plot of ***TotalBill V/S Tip*** of a restaurant.

````python
plt.scatter(
    x = tips_df['total_bill'],
    y = tips_df['tip'],
    marker = 'x'
)
plt.title('TotalBill V/S Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
````

Now, let's add another level of information on top of it.

````python
plt.scatter(
    x = tips_df['total_bill'],
    y = tips_df['tip'],
    marker = 'x',
    color = tips_df['sex'].apply(
        func = lambda x : 'hotpink' if x == 'Female' else 'dodgerblue'
    )
)
plt.title('TotalBill V/S Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
````

Now, we are representing one more information with the help of the colors of the markers, where ***HotPink*** color represents ***Female*** and ***DodgerBlue*** represents ***Male***.

Similarly, we can colorize it for ***Smoker*** and ***Non-Smoker***.

````python
plt.scatter(
    x = tips_df['total_bill'],
    y = tips_df['tip'],
    marker = 'x',
    color = tips_df['smoker'].apply(
        func = lambda x : 'orange' if x == 'Yes' else 'deepskyblue'
    )
)
plt.title('TotalBill V/S Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
````

But the above scatter plot is less readable as it doesn't show which colors represent smoker and non-smokers.  
Here is a away to deal with it, but a comprehesive approach...

````python
smokers = tips_df[tips_df['smoker'] == 'Yes']
non_smokers = tips_df[tips_df['smoker'] == 'No']
plt.scatter(
    x = smokers['total_bill'],
    y = smokers['tip'],
    marker = 'x',
    color = 'orange',
    label = 'Smoker'
)
plt.scatter(
    x = non_smokers['total_bill'],
    y = non_smokers['tip'],
    marker = 'x',
    color = 'deepskyblue',
    label = 'Non-Smoker'
)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
````

One more interesting thing about ***scatter*** method is ***s*** parameter. It allows us to increase or decrease size of the marker, which if we use smartly, we can have something interesting visualization like below.

````python
smokers = tips_df[tips_df['smoker'] == 'Yes']
non_smokers = tips_df[tips_df['smoker'] == 'No']
plt.scatter(
    x = smokers['total_bill'],
    y = smokers['tip'],
    marker = 'x',
    color = 'orange',
    label = 'Smoker',
    s = smokers['size']*10
)
plt.scatter(
    x = non_smokers['total_bill'],
    y = non_smokers['tip'],
    marker = 'x',
    color = 'deepskyblue',
    label = 'Non-Smoker',
    s = non_smokers['size']*10
)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
````

Now, we can see in the above visualization that we are representing information of ***Four-Dimensions*** in a single scatter plot.  
1. Total Bill
2. Tip
3. Smoker
4. Size (Number of Peoples)

Now, what if I want to plot the ***Runs Scored in the year 2015, 2016 and 2017, all in a single plot***. Think of it, will it be possible?

````python
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2015']
)
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2016']
)
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2017']
)
plt.title('Runs Scored in the Year 2015, 2016 and 2017')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
````

***No***, the way we tried to plot for all years in a single plot ***didn't work out***. It is not ***readable***.

Then what's the ***right way*** to do it.  
**Answer :** The first way to do it is ***Grouped Bar Chart***. Another way to do it is ***Stacked Bar Chart***.

#### **2. Grouped Bar Chart**

````python
bar_width = 0.8 / 3

plt.bar(
    x = np.arange(
        stop = batsman_season_records_df.shape[0]
    ) - bar_width,
    height = batsman_season_records_df['2015'],
    width = bar_width,
    label = '2015',
    color = 'indianred'
)
plt.bar(
    x = np.arange(
        stop = batsman_season_records_df.shape[0]
    ),
    height = batsman_season_records_df['2016'],
    width = bar_width,
    label = '2016',
    color = 'gold'
)
plt.bar(
    x = np.arange(
        stop = batsman_season_records_df.shape[0]
    ) + bar_width,
    height = batsman_season_records_df['2017'],
    width = bar_width,
    label = '2017',
    color = 'mediumorchid'
)
plt.title('Runs Scored in the Year 2015, 2016 and 2017')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.xticks(
    ticks = np.arange(
        stop = batsman_season_records_df.shape[0]
    ),
    labels = batsman_season_records_df['batsman']
)
plt.legend()
plt.show()
````

#### **3. Stacked Bar Chart**

1. The visualization in the above section has an alternative visualization known as ***Stacked Bar Chart***, where ***bars are stacked on each other***.  
2. Likewise Grouped Bar Chart, Matplotlib does not have easy method for this, instead, we have to ***play smart to make it happen*** as done in the case of Grouped Bar Chart.  
3. For Stacked Bar Chart, we will be playing smart with the help of ***bottom*** parameter of the ***bar()*** method which ***allows us shift the base of bars in Y-axis***.

````python
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2015'],
    bottom = 0,
    label = '2015',
    color = 'indianred'
)
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2016'],
    bottom = batsman_season_records_df['2015'],
    label = '2016',
    color = 'gold'
)
plt.bar(
    x = batsman_season_records_df['batsman'],
    height = batsman_season_records_df['2017'],
    bottom = batsman_season_records_df['2015'] + batsman_season_records_df['2016'],
    label = '2017',
    color = 'mediumorchid'
)
plt.title('Runs Scored in the Year 2015, 2016 and 2017')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.legend()
plt.show()
````

### üéØ **Pie Chart**  
To plot a Pie Chart you just need to call *matplotlib.pyplot.pie()* function.

Let's take a simple example.

````python
marks = [60, 70, 80, 90, 100]
subjects = ['English', 'Hindi', 'Science', 'Computer', 'Mathematics']
````

````python
plt.pie(
    x = marks,
    labels = subjects
)
plt.show()
````

Ummm, the colors are not cool. Can we do something?  
The answer is ***Yes***.

This below code will help you to check for all the colormaps available in Matplotlib.  
You can also check this link : [CMAPS](https://matplotlib.org/stable/users/explain/colors/colormaps.html)

````python
from matplotlib import colormaps
list(colormaps)
````

````python
cmap = plt.get_cmap('Purples')

plt.pie(
    x = marks,
    labels = subjects,
    colors = [
        cmap(i) for i in np.linspace(0, 1, len(marks))
    ]
)
plt.show()
````

Let's say I want to emphasize on the slice representing highest marks. How can I do that?  
You may give a try to ***explode*** parameter of the ***pie()*** function.

````python
cmap = plt.get_cmap('Purples')

plt.pie(
    x = marks,
    labels = subjects,
    colors = [
        cmap(i) for i in np.linspace(0, 1, len(marks))
    ],
    explode = [
        0 if x != max(marks) else 0.3 for x in marks
    ],
    shadow = True
)
plt.show()
````

**Problem :** All slices are looking almost same. How to differentiate or analyse?  
**Solution :** Use autopct to display percentage of each slice.

````python
cmap = plt.get_cmap('Purples')

plt.pie(
    x = marks,
    labels = subjects,
    colors = [
        cmap(i) for i in np.linspace(0, 1, len(marks))
    ],
    explode = [
        0 if x != max(marks) else 0.3 for x in marks
    ],
    shadow = True,
    autopct = '%0.2f%%'
)
plt.show()
````

We can actually save any visualizations as an image in our local device.

Let's save a visualization.

````python
np.random.seed(42)
data = np.random.randint(1, 100, 200)

plt.hist(
    x = data,
    bins = 'auto'
)
plt.title('Histogram of 200 Random Integers')
plt.xlabel('Interval')
plt.ylabel('Frequency')
plt.savefig('../Resources/Images/GGPlot Histogram of Random 200 Integers.png')
plt.show()
````

**Note :** Make sure to call the plt.savefig() function before calling plt.show() function else the visualization will be erased and you will see nothing in your saved image.
