### üéØ **Q01 : Line Plot of PM2.5 for Iran and China**

1. **Dataset:** Use the dataset containing air pollution levels of multiple countries and cities.

2. **Task Description:**

    You are required to draw a **line plot** comparing the air pollution level (PM2.5) between **Iran** and **China** over different **years**.

   * The **x-axis** should represent the `Year`.
   * The **y-axis** should represent the **sum of PM2.5** values for each year.
   * Include **only** the countries **Iran** and **China** in this plot.

3. **Expected Outcome:**

    The line plot should clearly illustrate how the PM2.5 levels have changed year by year for **Iran** and **China**, allowing for a visual comparison of pollution trends over time.

```python
iran_df = pollution_df[pollution_df.Country == 'Iran'][['PM2.5', 'Year']].groupby('Year').sum().reset_index()
china_df = pollution_df[pollution_df.Country == 'China'][['PM2.5', 'Year']].groupby('Year').sum().reset_index()
```

```python
plt.plot(
    iran_df['Year'],
    iran_df['PM2.5'],
    label = 'Iran',
    marker = 'x'
)
plt.plot(
    china_df['Year'],
    china_df['PM2.5'],
    label = 'China',
    marker = 'o'
)
plt.title('PM2.5 Value Comparison of Iran & China')
plt.xlabel('PM2.5')
plt.ylabel('Year')
plt.legend()
plt.show()
```

### üéØ **Q09 : Quarter-wise Multi-Bar Chart of Product Sales**

1. **Task Description:**

    Create a **multi-bar chart** (grouped bar chart) to represent the **sales data of all products, grouped by quarter**.

    Your task involves:

    1. **Aggregate monthly sales into quarters:**

       * **Q1** ‚Üí January, February, March
       * **Q2** ‚Üí April, May, June
       * **Q3** ‚Üí July, August, September
       * **Q4** ‚Üí October, November, December

    2. **Plot a grouped bar chart**, where:

       * The **x-axis** represents the quarters (`Q1`, `Q2`, `Q3`, `Q4`)
       * For each quarter, display **bars for all products**, grouped together

    3. Add the following:

        * **Legend** indicating product names
        * **Title** and **axis labels** as appropriate

2. **Expected Outcome:**

    A cleanly grouped bar chart showing each product's quarterly sales performance, visually comparing sales across quarters.

```python
from math import ceil

sales_df['quarter'] = sales_df['month_number'].apply(
    func = lambda x : ceil(x / 3)
).apply(
    func = lambda x : 'Q' + str(x)
)
```

```python
facecream = sales_df[['facecream', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'facecream' : 'sum'
    }
)
facewash = sales_df[['facewash', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'facewash' : 'sum'
    }
)
toothpaste = sales_df[['toothpaste', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'toothpaste' : 'sum'
    }
)
bathingsoap = sales_df[['bathingsoap', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'bathingsoap' : 'sum'
    }
)
shampoo = sales_df[['shampoo', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'shampoo' : 'sum'
    }
)
moisturizer = sales_df[['moisturizer', 'quarter']].groupby(
    by = 'quarter'
).agg(
    func = {
        'moisturizer' : 'sum'
    }
)
```

```python
bar_width = 0.8 / 6

plt.figure(figsize = (16, 8))
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ) - 2 * bar_width,
    height = facecream['facecream'],
    width = bar_width,
    label = 'Face Cream'
)
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ) - bar_width,
    height = facewash['facewash'],
    width = bar_width,
    label = 'Face Wash'
)
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ),
    height = toothpaste['toothpaste'],
    width = bar_width,
    label = 'Toothpaste'
)
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ) + bar_width,
    height = bathingsoap['bathingsoap'],
    width = bar_width,
    label = 'Bathing Soap'
)
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ) + 2 * bar_width,
    height = shampoo['shampoo'],
    width = bar_width,
    label = 'Shampoo'
)
plt.bar(
    x = np.arange(
        stop = facecream.shape[0]
    ) + 3 * bar_width,
    height = moisturizer['moisturizer'],
    width = bar_width,
    label = 'Moisturizer'
)
plt.title('Quarterly Grouped Bar Chart of Products')
plt.xlabel('Quarter')
plt.ylabel('Units Sold')
plt.legend(title = 'Products')
plt.xticks(
    ticks = np.arange(
        stop = facecream.shape[0]
    ),
    labels = facecream.index
)
plt.show()
```

### üéØ **Q10 : Quarter-Wise Stacked Bar Chart for Each Product**

1. **Dataset Requirement:** Use the dataset that contains **monthly sales data** for each product. The dataset should include:

* A column for **Product Name**
* Sales columns for each **month**: `"January"` to `"December"`

2. **Task Description:**

    Create a **stacked bar chart** that visualizes **quarter-wise sales** for each product.

    Your task involves:

    1. **Aggregate monthly sales into quarters:**

    * **Q1** ‚Üí January, February, March
    * **Q2** ‚Üí April, May, June
    * **Q3** ‚Üí July, August, September
    * **Q4** ‚Üí October, November, December

    2. **Plot a stacked bar chart** where:

    * The **x-axis** represents each **product**
    * Each bar (per product) is divided into **stacked segments** representing **sales for Q1, Q2, Q3, and Q4**

    3. **Include the following visual enhancements:**

    * **Legend** to distinguish between quarters
    * **Title**, **x-axis**, and **y-axis labels**

3. **Expected Outcome:**

    A stacked bar chart showing how each product performed in different quarters, stacked on top of each other for cumulative visibility.

```python
products_group = sales_df.drop(
    columns = [
        'month_number',
        'total_profit',
        'total_units'
    ]
).groupby(
    by = 'quarter'
)
```

```python
q1 = products_group.get_group('Q1').sum().drop('quarter')
q2 = products_group.get_group('Q2').sum().drop('quarter')
q3 = products_group.get_group('Q3').sum().drop('quarter')
q4 = products_group.get_group('Q4').sum().drop('quarter')
```

```python
plt.figure(figsize = (16, 8))
plt.bar(
    x = np.arange(
        stop = len(q1)
    ),
    height = q1,
    bottom = 0,
    label = 'Q1'
)
plt.bar(
    x = np.arange(
        stop = len(q2)
    ),
    height = q2,
    bottom = q1,
    label = 'Q2'
)
plt.bar(
    x = np.arange(
        stop = len(q3)
    ),
    height = q3,
    bottom = q1 + q2,
    label = 'Q3'
)
plt.bar(
    x = np.arange(
        stop = len(q4)
    ),
    height = q4,
    bottom = q1 + q2 + q3,
    label = 'Q4'
)
plt.title('Stacked Bar Chart')
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.legend(
    title = 'Quarter'
)
plt.xticks(
    ticks = np.arange(
        stop = len(q1)
    ),
    labels = q1.index.str.upper()
)
plt.show()
```
