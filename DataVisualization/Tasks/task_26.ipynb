### üéØ **Q04 : Categorical and Regression Analysis using Taxis Dataset**

1. **Dataset:** Seaborn `taxis` Dataset

2. **Task Description:**

   * Create a **categorical estimate plot** showing the total fare (`total`) for each payment type (`payment`).
   * Create a **regression plot** to visualize the relationship between ride time and total fare.
   * Use the pickup and dropoff time columns to derive the ride duration.
   * Create one regression plot with **hue based on payment type**.
   * Create another regression plot with **hue based on taxi color**.
   * Observe and compare the visual patterns across the plots.

3. **Expected Outcome:**
   A set of plots that illustrate how total fare varies across payment types and how ride duration relates to fare under different categorical distinctions.

```python
from datetime import timedelta
taxis_df['ride_time'] = taxis_df['dropoff'] - taxis_df['pickup']
taxis_df['ride_time'] = taxis_df['ride_time'].apply(timedelta.total_seconds)

sns.lmplot(data=taxis_df, x='ride_time', y='total', aspect=2)
plt.show()
```

```python
sns.lmplot(data=taxis_df, x='ride_time', y='total',
           hue='payment', col='payment')
plt.show()
```

```python
sns.lmplot(data=taxis_df, x='ride_time', y='total',
           hue='color', col='color')
plt.show()
```

### üéØ **Q05 : Strip Plot and Swarm Plot of Blood Pressure by Gender**

1. **Dataset:** Insurance Dataset *(CSV provided)*

2. **Task Description:**

   * Create a **strip plot** to visualize the relationship between `gender` and `bloodpressure`.
   * Use the `smoker` column as the **hue** parameter in the strip plot.
   * Create a **swarm plot** for the same variables: `gender` and `bloodpressure`.
   * Use the `smoker` column as the **hue** parameter in the swarm plot.
   * Add appropriate titles to both charts.

3. **Expected Outcome:**
   Two categorical distribution plots (strip and swarm) showing blood pressure variations across gender, distinguished by smoker status.


**Note :** `y=1.02` is used to provide a vertical gap between the visualization and the title.

```python
g = sns.catplot(data=ins_df, x='gender', y='bloodpressure', kind='strip',
            hue='smoker')
g.figure.suptitle('Categorical Distribution of Blood Pressure', y=1.02)
plt.show()
```

### üéØ **Q10 : Pair Grid Visualization of Insurance Dataset by Diabetic Status**

1. **Dataset:** Insurance Dataset

2. **Task Description:**

   * Create a **pair grid** using the insurance dataset.
   * Use the `diabetic` column as the **hue** parameter.
   * Display **box plots** on the diagonal.
   * Display **scatter plots** on the upper triangle.
   * Display **KDE plots** on the lower triangle.
   * Add an appropriate title to the visualization.

3. **Expected Outcome:**
   A comprehensive pair grid showing distributions and relationships among variables in the insurance dataset, differentiated by diabetic status.


```python
g = sns.PairGrid(data=ins_df, hue='diabetic',
                 vars=['age', 'bmi', 'bloodpressure', 'children', 'claim'])
g.map_diag(sns.boxplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.figure.suptitle('Pair Grid Visualization of Insurance Dataset by Diabetic Status', y=1.02)
plt.show()
```

### üéØ **Q12 : Joint Grid of Age vs Claim**

1. **Dataset:** Insurance Dataset

2. **Task Description:**

   * Create a **joint grid** visualization.
   * Set the X-axis to represent `age`.
   * Set the Y-axis to represent `claim`.
   * Include appropriate titles and axis labels.

3. **Expected Outcome:**
   A joint grid visualizing the relationship between age and claim values.


```python
g = sns.JointGrid(data=ins_df, x='age', y='claim')
g.plot(joint_func=sns.scatterplot,
       marginal_func=sns.kdeplot)
plt.show()
```
