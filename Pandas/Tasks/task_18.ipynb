### **üéØ Q1: Team-wise On-Target Percentage in Football Dataset**  

Compute the **percentage of total attempts that were on target** for each team in the dataset.  

1. **Instructions:**  
   1. Identify the **total number of unique teams** that participated in the tournament.  
   2. For each team:  
      - Filter the dataset to include only matches where the team played.  
      - Calculate the **total number of attempts** and **total on-target attempts**.  
      - Compute the percentage using the formula:  
        $$
        \text{On-Target Percentage} = \left(\frac{\text{Total On-Target Attempts}}{\text{Total Attempts}}\right) \times 100
        $$
      - Round the result to **2 decimal places**.  
   3. Store the values in a **Python dictionary**, where the **keys** are team names and the **values** are the computed percentages.  
   4. Sort the dictionary by **percentage values in descending order** and display the final result.  

2. **Expected Output Format:**  
   - A **Python dictionary** containing team names as **keys** and their **on-target percentage** as values, **sorted** in descending order.  
   - Example Output:  
     ```python
     {'Costa Rica': 54.55, 'Cameroon': 51.85, 'Ecuador': 48.15, ... , 'Canada': 17.65}
     ```  

````python
# Solution - Question 01

def prepare_dict(teams, df):
    output = {}
    for i in teams:
        mask = df['Team'] == i
        total_attempts = df[mask]['Total Attempts'].sum()
        on_target_attempts = df[mask]['On Target'].sum()
        percentage = round((on_target_attempts / total_attempts) * 100, 2)
        output[i] = percentage
    return dict(sorted(output.items(), key=lambda item: item[1], reverse=True))

teams = fifa_df['Team'].unique()
prepare_dict(
    teams = teams,
    df = fifa_df
)
````

3. **‚úÖ Expected Output Format:**  
   - A **sorted DataFrame** showing each team's **performance statistics**.  
   - Display a **few rows** to verify correctness.

````python
ipl_df.replace(
    to_replace = {
        'Delhi Daredevils' : 'Delhi Capitals',
        'Kings XI Punjab' : 'Punjab Kings',
        'Rising Pune Supergiants' : 'Rising Pune Supergiant'
    },
    inplace = True
)
````

````python
all_teams = np.unique(
    np.append(
        ipl_df['Team1'].unique(),
        ipl_df['Team2'].unique()
    )
)
matches_df = ipl_df[~ipl_df.WinningTeam.isna()]
````

````python
data = []
for team in all_teams:
    match_count = matches_df[(matches_df['Team1'] == team) | (matches_df['Team2'] == team)].shape[0]
    winning_count = matches_df[(matches_df['WinningTeam'] == team)].shape[0]
    winning_percentage = round((winning_count / match_count) * 100, 2)

    home_matches_count = matches_df[(matches_df['Team1'] == team)].shape[0]
    home_win_count = matches_df[(matches_df['Team1'] == team) & (matches_df['WinningTeam'] == team)].shape[0]
    home_winning_percentage = round((home_win_count / home_matches_count) * 100, 2)

    away_matches_count = matches_df[(matches_df['Team2'] == team)].shape[0]
    away_win_count = matches_df[(matches_df['Team2'] == team) & (matches_df['WinningTeam'] == team)].shape[0]
    away_winning_percentage = round((away_win_count / away_matches_count) * 100, 2)
    
    data.append(
        [
            team,
            match_count,
            winning_percentage,
            home_winning_percentage,
            away_winning_percentage
        ]
    )

team_performance_df = pd.DataFrame(
    data = data,
    columns = ['Team Name', 'Matches Played', 'Win %', 'Home Win %', 'Away Win %']
)
````

### **üéØ Q9: IPL Points Table**  

Create a function **`point_table(season)`** that generates the **IPL Points Table** for a given season.  

1. **üìå Function Behavior:**  
   - The function takes **one parameter**:  
     ```python
     point_table(season: str)
     ```  
   - The `season` must be one of the following valid season names:  
     ```python
     ['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', 
      '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
     ```
  
2. **‚úÖ Points System:**  
   - **Win** ‚Üí +2 Points  
   - **Loss** ‚Üí +0 Points  
   - **No Result** ‚Üí +1 Point per team  

3. **üìä Required DataFrame Structure:**  
    | **TeamName**        | **MatchesPlayed** | **MatchesWon** | **NoResult** | **Points** |
    |----------------------|------------------|---------------|------------|----------|
    | Gujarat Titans      | 16               | 12            | 0          | 24       |
    | Rajasthan Royals    | 17               | 10            | 0          | 20       |  

4. **üìå Sorting Criteria:**  
   1. **Sort by "Points" in descending order** (highest points first).  
   2. **If teams have the same points, sort by "TeamName" in ascending order** (A ‚Üí Z).  

5. **‚ö†Ô∏è Additional Notes:**  
   - **Exclude abandoned seasons** if any.  
   - **Make "TeamName" the index** in the DataFrame.  

üöÄ Call Example:  
```python
point_table('2022')
```

````python
def point_table(season : str):
    season_matches = ipl_df[ipl_df.Season == season]
    all_teams = np.unique(
        np.append(
            season_matches['Team1'].unique(),
            season_matches['Team2'].unique()
        )
    )
    points_data = []
    for team in all_teams:
        matches_played = season_matches[(season_matches.Team1 == team) | (season_matches.Team2 == team)].shape[0]
        matches_won = season_matches[season_matches.WinningTeam == team].shape[0]
        no_result_matches = season_matches[
            season_matches.WinningTeam.isna() & 
            ((season_matches.Team1 == team) | (season_matches.Team2 == team))
            ].shape[0]
        team_points = (matches_won * 2) + (no_result_matches * 1)
        points_data.append(
            [
                team,
                matches_played,
                matches_won,
                no_result_matches,
                team_points
            ]
        )
    df = pd.DataFrame(
        data = points_data,
        columns = [
            'TeamName',
            'MatchesPlayed',
            'MathcesWon',
            'NoResult',
            'Points'
        ],
    ).sort_values(by = ['Points', 'TeamName'], ascending = [False, True]).set_index(keys = 'TeamName')

    return df
````

### **üéØ Q10: IPL Points Table (Extended)**  

Extend the **IPL Points Table** from the previous question by adding an **extra column**: `"SeasonPosition"`.  

1. **üìå Column: `SeasonPosition` Rules**  
    After **sorting by**:  
    1. **"Points"** (descending order)

Assign the `"SeasonPosition"` as follows:  

2. **üèÜ Top 4 Teams**  
    | Position | Condition |
    |----------|-----------|
    | **Winner** | Team that won the Final |
    | **Runner** | Team that lost the Final |
    | **3** | Losing team in **Qualifier 2** |
    | **4** | Losing team in **Eliminator** |

3. **üìä Other Teams**  
   - All remaining teams will have `"SeasonPosition"` **equal to their rank**, using `rank()`.

4. **‚úÖ Updating Specific Cells in a DataFrame**  
   - Use **`df.at[row_index, col_label] = value`** to modify individual cells.  
   - **Avoid SettingWithCopyWarning** by working on a copy of the DataFrame:

5. **üìä Expected DataFrame Format**  
    | **TeamName**        | **MatchesPlayed** | **MatchesWon** | **NoResult** | **Points** | **SeasonPosition** |
    |----------------------|------------------|---------------|------------|----------|------------------|
    | Gujarat Titans      | 16               | 12            | 0          | 24       | **Winner**       |

    ```plaintext
    Gujarat Titans          16         12         0       24       Winner

    ...
    ```

7. **üìå Notes:**  
   - The **function should return the complete table** with `"SeasonPosition"`.  
   - Ensure the `"SeasonPosition"` column is updated **correctly** based on match results.  
   - Use **`rank()`** for non-top-4 teams.  

üöÄ **Call Example:**  
```python
points_table('2022')
```

````python
def get_loosing_team(df):
    if df.WinningTeam == df.Team1:
        return df.Team2
    else:
        return df.Team1
````

````python
def point_table_extension(season : str):
    season_matches = ipl_df[ipl_df.Season == season].copy()
    table = point_table(season = season)

    table['SeasonPosition'] = table.Points.rank(
        method = 'first',
        ascending = False
    ).astype('object')

    winning_team = season_matches.WinningTeam[season_matches.MatchNumber == 'Final'].values[0]
    table.at[winning_team, 'SeasonPosition'] = 'Winner'

    season_matches['LosingTeam'] = season_matches.apply(func = get_loosing_team, axis=1)

    runner_team = season_matches.LosingTeam[season_matches.MatchNumber == 'Final'].values[0]
    table.at[runner_team, 'SeasonPosition'] = 'Runner'

    third_team = season_matches.LosingTeam[season_matches.MatchNumber == 'Qualifier 2'].values[0]
    table.at[third_team, 'SeasonPosition'] = 3

    fourth_team = season_matches.LosingTeam[season_matches.MatchNumber == 'Eliminator'].values[0]
    table.at[fourth_team, 'SeasonPosition'] = 4

    return table
````
