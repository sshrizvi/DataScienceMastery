### **🎯 Q3: Handling Missing Values in DataFrame**  

1. **Select Rows with NaN Values:**  
    Display all rows where either **`age`** or **`visits`** column has missing (NaN) values.  

2. **Fill NaN Values:**  
    Replace missing values in both **`age`** and **`priority`** columns with their respective **mode** (most frequent) value.  

````python
# Solution - Question 03

mask = (df['age'].isna()) | (df['visits'].isna())
df.loc[mask]
````

````python
modes = {
    'birds' : df['birds'].mode().iloc[0],
    'age' : df['age'].mode().iloc[0],
    'visits' : df['visits'].mode().iloc[0],
    'priority' : df['priority'].mode().iloc[0],
}

df.fillna(value=modes, inplace=True)
df
````

### **🎯 Q8: Team1 vs Team2 Dashboard**  

Create a function `team1_vs_team2(team1, team2)` that takes two team names as input and provides the following details:  
1. **Win-Loss Record** between the two teams.  
   - Show how many matches each team has won.  

2. **Player with Most Player of the Match Awards** in matches between these two teams.  

3. **Example:**  
   ```python
   team1_vs_team2('Kolkata Knight Riders', 'Chennai Super Kings')
   ```  

4. **Expected Output:**  
   - **Win-Loss Record**: Display total wins for each team.  
   - **Player with Most Player of the Match Awards**: Identify the player with the highest number of Player of the Match awards in matches between these teams.  

````python
# Solution - Question 08

def team1_vs_team2(team1, team2):
    team1_mask = (ipl_df['Team1'] == team1) | (ipl_df['Team2'] == team1)
    team2_mask = (ipl_df['Team2'] == team2) | (ipl_df['Team1'] == team2)
    match_records = ipl_df[team1_mask & team2_mask]
    team1_wins = match_records['WinningTeam'].value_counts().loc[team1]
    team2_wins = match_records['WinningTeam'].value_counts().loc[team2]
    print('{} wins {} matches.'.format(team1, team1_wins))
    print('{} wins {} matches.'.format(team2, team2_wins))
    freq_potm = match_records['Player_of_Match'].value_counts().sort_values(ascending=False).index[0]
    print('Most frequent Player of the Match is {}.'.format(freq_potm))
    

team1_vs_team2('Kolkata Knight Riders', 'Chennai Super Kings')
````

### **🎯 Q10: Average Winning Margin of Mumbai Indians (2011 Season)**  

Calculate the **average winning margin** for **Mumbai Indians** in the **2011 IPL season**.  

1. **Instructions:**  
   1. Filter matches played by **Mumbai Indians** during the **2011 season**.  
   2. Consider the winning margin only from the matches won by **Mumbai Indians**.  
   3. Compute the **average winning margin** (both runs and wickets).  

2. **Expected Output:**  
   - The average winning margin in terms of **runs** and **wickets** for the **Mumbai Indians** in the 2011 season.  

````python
season_mask = ipl_df['Season'] == '2011'
winning_team_mask = ipl_df['WinningTeam'] == 'Mumbai Indians'
margin_df = ipl_df[season_mask & winning_team_mask][['WonBy', 'Margin']]
won_by = margin_df['WonBy'].unique()

for i in won_by:
    won_by_category_mask = margin_df['WonBy'] == i
    won_by_category_data = margin_df[won_by_category_mask]['Margin']
    print('Average Margin of Winning by {} for {} is {}.'.format(i, 'Mumbai Indians', won_by_category_data.mean()))
````
