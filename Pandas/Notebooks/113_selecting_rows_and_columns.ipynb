```python
dict_data = {
    'NAME' : ['Shujaat', 'Abbas', 'Ali', 'Chandra', 'Verma'],
    'IQ' : [100, 120, 140, 160, 180],
    'MARKS' : [200, 220, 240, 260, 280],
    'PACKAGE' : [60, 70, 80, 90, 100]
}
students_df = pd.DataFrame(data = dict_data)
students_df.set_index('NAME', inplace=True)
students_df
```

### üöÄ **Selecting Rows**

#### üéØ **1. Using `iloc[]`**

##### **a. Single Row**

```python
ipl_df.iloc[0]
```

```python
type(ipl_df.iloc[0])
```

##### **b. Multiple Rows using Slicing**

```python
movies_df.iloc[0:10]
```

##### **c. Multiple Rows using Fancy Indexing**

```python
movies_df.iloc[[1, 5, 940]]
```

##### **d. Multiple Rows using Boolean Array**

```python
mask = movies_df.index % 100 == 0
movies_df.iloc[mask]
```

#### üéØ **2. Using `loc[]`**

##### **a. Single Row**

```python
students_df.loc['Shujaat']
```

##### **b. Multiple Rows using Fancy Indexing**

```python
students_df.loc[['Shujaat', 'Ali', 'Verma']]
```

##### **c. Multiple Rows using Slicing**

```python
students_df.loc['Ali':'Chandra']
```

```python
students_df.loc['Shujaat':'Verma':2]
```

##### **d. Multiple Rows using Boolean Array**

```python
mask = students_df.index == 'Ali'
students_df.loc[mask]
```

##### **e. Multiple Rows using Callable Function**

```python
students_df.loc[lambda x : x['PACKAGE'] > x['PACKAGE'].mean()]
```

### üöÄ **Selecting Both Rows and Columns**

#### üéØ **1. Using `iloc[row, column]`**

```python
ipl_df.iloc[0:5, 0:3]
```

#### üéØ **2. Using `loc[row, column]`**

```python
students_df.loc['Shujaat':'Ali', 'PACKAGE']
```
