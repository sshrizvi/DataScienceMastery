```python
pd.merge(
    left = students,
    right = november,
    how = 'right',
    on = 'student_id'
)
```

### üöÄ **Questions on Joins**

#### üéØ **Q1. Find the total revenue generated.**

```python
total_registrations = pd.concat(
    objs = [
        november,
        december
    ],
    axis = 0,
    ignore_index = True
)

total_revenue = pd.merge(
    left = total_registrations,
    right = courses,
    how = 'inner', 
    on = 'course_id'
)['price'].sum()

print(
    'Total Revenue Generated : {}'.format(
        total_revenue
    )
)
```

#### üéØ **Q2. Find Month by Month Revenue.**

```python
total_registrations = pd.concat(
    objs = [
        november,
        december
    ],
    axis = 0,
    keys = [
        'November',
        'December'
    ]
)

for i in total_registrations.index.levels[0]:
    month_revenue = pd.merge(
        left = total_registrations.loc[i],
        right = courses,
        how = 'inner',
        on = 'course_id'
    )['price'].sum()
    print(
        'Revenue for Month {} is Rs. {}'.format(
            i, month_revenue
        )
    )
```

#### üéØ **Q3. Print the Registration Table (Name, Course, Price)**

```python
total_registrations = pd.concat(
    objs = [
        november,
        december
    ],
    axis = 0,
    ignore_index = True
)

pd.merge(
    left = total_registrations,
    right = students,
    how = 'inner',
    on = 'student_id'
).merge(
    right = courses,
    how = 'inner',
    on = 'course_id'
)[
    ['name', 'course_name', 'price']
]
```

#### üéØ **Q4. Plot bar chart of revenue for each course.**

```python
total_registrations = pd.concat(
    objs = [
        november,
        december
    ],
    axis = 0,
    ignore_index = True
)

pd.merge(
    left = total_registrations,
    right = courses,
    how = 'inner',
    on = 'course_id'
)[['course_name', 'price']].groupby(
    by = 'course_name'
).sum().plot(
    kind = 'bar'
)
```

### üöÄ **Questions on IPL DataSet**

#### üéØ **Q1. Find the TOP 3 Stadiums with highest SIXES/MATCHES ratio.**

```python
def sixes_matches_ratio(group):
    match_count = group['id'].count()
    sixes_count = group['sixes_count'].sum()
    return (sixes_count/match_count)
```

```python
match_run_count = deliveries.groupby(
    by = 'match_id'
)['batsman_runs'].value_counts().reset_index().rename(
    columns = {
        'count' : 'sixes_count'
    }
)

sixes = match_run_count[
    match_run_count['batsman_runs'] == 6
]

venue_group = pd.merge(
    left = matches,
    right = sixes,
    left_on = 'id',
    right_on = 'match_id'
).groupby(
    by = 'venue'
)

venue_group.apply(
    func = sixes_matches_ratio,
    include_groups = False
).sort_values(
    ascending = False
).head(3)
```

#### üéØ **Q2. Find ORANGE CAP holder for all the seasons.**

```python
def orange_cap(group):
    return group.sort_values(
        by = 'batsman_runs',
        ascending = False
    ).head(1)
```

```python
temp_df = pd.merge(
    left = matches,
    right = deliveries,
    left_on = 'id',
    right_on = 'match_id'
)

temp_group = temp_df.groupby(
    by = [
        'season',
        'batsman'
    ]
)

total_runs = temp_group['batsman_runs'].sum().reset_index()

total_runs.groupby(
    by = 'season'
).apply(
    func = orange_cap,
    include_groups = False
)
```
