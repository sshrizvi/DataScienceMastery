##### üéØ **Q1. Find the last match played by Virat Kohli in Delhi.**

````python
ipl_df['All Players'] = ipl_df['Team1Players'] + ipl_df['Team2Players']
````

````python
def did_kohli_played(players):
    return 'V Kohli' in players
````

````python
ipl_df['did_kohli_play'] = ipl_df['All Players'].apply(func=did_kohli_played)
````

````python
ipl_df[(ipl_df['City'] == 'Delhi') & (ipl_df['did_kohli_play'] == True)].drop_duplicates(
    subset=['City', 'did_kohli_play'],
    keep='first'
)
````

#### üì¶ **16. DataFrame.apply()**  

**Documentation Link :** [Link](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html#pandas-dataframe-apply)

**Description :** Apply a function along an axis of the DataFrame.

````python
def euclidean(row):
    p1 = row['Point1']
    p2 = row['Point2']
    return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5
````

````python
points['Distance'] = points.apply(
    func=euclidean,
    axis=1
)
points
````
