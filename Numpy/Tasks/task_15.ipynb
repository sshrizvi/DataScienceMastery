### ðŸŽ¯ **Q1: Create a Random 4x3 Matrix and Perform Sorting Tasks**  

1. **Problem Statement**:  
   Write a program to create a `4x3` matrix with random integer values between 0 and 100. Perform various sorting operations on this matrix as outlined below.  

2. **Inputs**:  
   - A randomly generated matrix of shape `(4, 3)` with integers in the range `[0, 100]`.  

3. **Tasks**:  
   i. **Sort the Matrix Row-wise**:  
      Use the `np.sort()` function to sort the elements in each row of the matrix in ascending order.  

   ii. **Sort Based on Values in the Second Column**:  
       Sort the rows of the matrix based on the values in the second column (column index `1`).  

   iii. **Sort Based on the Maximum Value in Each Row**:  
        Rearrange the rows of the matrix in descending order of the maximum value in each row.  

   iv. **Sort Based on All Elements in Ascending Order**:  
       Flatten the matrix and sort all elements in ascending order, then reshape back into a `4x3` matrix.  

4. **Output Format**:  
   - A matrix for each sorting operation.  

5. **Example**:  

   - Input Matrix:  
     ```bash
     arr = 
     [[92 90 74]
      [ 6 63 93]
      [15 93 96]
      [70 60 48]]
     ```  

   - **Output**:  
     i. Row-wise Sorted:  

        ```bash
        [[74 90 92]
         [ 6 63 93]
         [15 93 96]
         [48 60 70]]
        ```  

     ii. Sorted by Second Column: 

        ```bash
        [[70 60 48]
         [ 6 63 93]
         [92 90 74]
         [15 93 96]]
        ```

     iii. Sorted by Row-wise Max Value (Descending):  

        ```bash
        [[15 93 96]
         [ 6 63 93]
         [92 90 74]
         [70 60 48]]
        ```

     iv. Sorted by All Elements:  

        ```bash
        [[ 6 15 48]
         [60 63 70]
         [74 90 92]
         [93 93 96]]
        ```

6. **Additional Notes**:  
   - Use NumPy for matrix creation and sorting operations.  
   - Ensure the program handles edge cases, such as duplicate values or edge values like `0` and `100`.  

````python
# Solution - Question 01

# Part (i)

arr =  [[92, 90, 74],
        [6 , 63, 93],
        [15, 93, 96],
        [70, 60, 48]]
array = np.array(arr)

print(' Sorted Array '.center(60, '-'))
print()
print(np.sort(array))
print()

# Part (ii)

arr =  [[92, 90, 74],
        [6 , 63, 93],
        [15, 93, 96],
        [70, 60, 48]]
array = np.array(arr)

print(' Sorted by Second Column '.center(60, '-'))
print()
sorted_indices = np.argsort(array[:,1])
print(array[sorted_indices])
print()

# Part (iii)

arr =  [[92, 90, 74],
        [6 , 63, 93],
        [15, 93, 96],
        [70, 60, 48]]
array = np.array(arr)

print(' Sorted by Row-wise Max Value (Descending) '.center(60, '-'))
print()
sorted_indices = np.argsort(np.max(array, axis=1))[::-1]
print(array[sorted_indices])
print()

# Part (iv)

arr =  [[92, 90, 74],
        [6 , 63, 93],
        [15, 93, 96],
        [70, 60, 48]]
array = np.array(arr)

print(' Sorted by All Elements '.center(60, '-'))
print()
print(np.sort(array, axis=None).reshape(array.shape))
````

   iv. **Sort the Array by Total Marks**:  
       Sort the array in descending order based on the total marks column. Display the top 2 rows from this sorted array.  

4. **Output Format**:  
   - Updated array after each task.  
   - Final sorted array and the top 2 rows based on total marks.  

5. **Example**:  

   - **Initial Array**:  
     ```bash
     [[13, 10,  9, 33],

      [82,  9, 29, 78],
      [67, 61, 59, 36]]
     ```  

   - **After Adding Extra Subject Marks**:  
     ```bash
     [[13, 10,  9, 33, 41],
      [63, 46, 90, 42, 87],
      [39, 76, 13, 29, 72],
      [82,  9, 29, 78, 36],
      [67, 61, 59, 36, 92]]
     ```  

   - **After Adding Two New Students**:  
     ```bash
     [[13, 10,  9, 33, 41],

      [39, 76, 13, 29, 72],
      [82,  9, 29, 78, 36],
     ```  

### ðŸŽ¯ **Q7: Find and Remove Minimum and Maximum Elements from an Array**

1. **Problem Statement**:  
   Write a program to find the minimum and maximum elements of a given NumPy array. Once identified, remove these elements from the array.

2. **Inputs**:  
   - A 1D NumPy array containing random integers.  
   Example:  
   ```python
   import numpy as np
   np.random.seed(400)
   arr = np.random.randint(100, 1000, 200).reshape((1, 200))
   ```

3. **Task**:  
   - Identify the **minimum** and **maximum** elements in the array.  
   - Remove these elements from the array while retaining the remaining structure.  
   - If there are duplicates of the minimum or maximum value, remove **only one occurrence** of each.

4. **Output Format**:  
   - A modified NumPy array without the minimum and maximum elements.  

5. **Example**:

   - **Input**:  
     ```bash
     arr = [345, 789, 120, 654, 120, 789]
     ```
   - **Process**:  
     - Minimum = 120  
     - Maximum = 789  
     - Remove one instance of each.  
   - **Output**:  
     ```bash
     [345, 789, 654, 120]
     ```

6. **Additional Notes**:  
   - Ensure the program works efficiently for large arrays.  
   - Use NumPy functions to perform the operations.  
   - The order of elements in the resulting array should remain unchanged except for the removed values.

````python
# Solution - Question 07

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))

min_val = np.min(arr)
max_val = np.max(arr)
new_arr = np.delete(arr, np.where(arr == min_val), axis=1)
output = np.delete(new_arr, np.where(new_arr == max_val), axis=1)
output
````

### ðŸŽ¯ **Q8: Limit Array Elements Between 100 and 200, Sort, and Perform Cumulative Sum**

1. **Problem Statement**:  
   You are given a NumPy array with elements in the range `[100, 700]`. Write a program to:  
   - Restrict the elements of the array such that values **less than 100** are replaced with `100` and values **greater than 200** are replaced with `200`.  
   - Sort the updated array in ascending order.  
   - Compute the cumulative sum of the sorted array.

2. **Inputs**:  
   - A 1D NumPy array with values in the range `[100, 700]`.

3. **Tasks**:  
   - Replace all elements less than `100` with `100` and all elements greater than `200` with `200`.  
   - Sort the resulting array.  
   - Calculate the cumulative sum of the sorted array using `np.cumsum()`.

4. **Output Format**:  
   - The updated array after limiting the values.  
   - The sorted array.  
   - The cumulative sum of the sorted array.

5. **Example**:

   - **Input**:  
     ```bash
     arr = [150, 250, 180, 700, 120, 90, 200, 300]
     ```
   - **Process**:  
     - Limit values: `[150, 200, 180, 200, 120, 100, 200, 200]`  
     - Sort: `[100, 120, 150, 180, 200, 200, 200, 200]`  
     - Cumulative Sum: `[100, 220, 370, 550, 750, 950, 1150, 1350]`  
   - **Output**:  
     ```bash
     Limited Array: [150, 200, 180, 200, 120, 100, 200, 200]
     Sorted Array: [100, 120, 150, 180, 200, 200, 200, 200]
     Cumulative Sum: [100, 220, 370, 550, 750, 950, 1150, 1350]
     ```

6. **Additional Notes**:  
   - Use NumPy functions for efficient manipulation of the array.  
   - Ensure the cumulative sum is displayed as a NumPy array.

````python
# Solution - Question 08

arr = np.array([150, 250, 180, 700, 120, 90, 200, 300])

clipped_array = np.clip(arr, a_min=100, a_max=200)
print(f'Limited Array : {clipped_array}')

sorted_array = np.sort(clipped_array)
print(f'Sorted Array : {sorted_array}')

cumsum_array = np.cumsum(sorted_array)
print(f'Cumulative Sum : {cumsum_array}')
````

### ðŸŽ¯ **Q9: Analyze Percentiles in a Given Array**

1. **Problem Statement**:  
   You are given a 1D NumPy array with elements in the range `[0, 1]`. Write a program to:  
   - Round off each element of the array to **3 decimal places**.  
   - Verify if the **0th percentile** of the array equals its **minimum value**.  
   - Verify if the **100th percentile** of the array equals its **maximum value**.  
   - Compute the difference between the **51st percentile** and the **50th percentile** values of the array.

2. **Inputs**:  
   - A 1D NumPy array where elements belong to the range `[0, 1]`.

3. **Tasks**:  
   - Round off all elements in the array to 3 decimal places using NumPy.  
   - Find the 0th percentile and verify if it matches the minimum value of the array.  
   - Find the 100th percentile and verify if it matches the maximum value of the array.  
   - Calculate the values at the 50th and 51st percentiles and find their difference.

4. **Output Format**:  
   - Rounded array with elements up to 3 decimal places.  
   - Validation results for:  
     - 0th percentile vs minimum value.  
     - 100th percentile vs maximum value.  
   - The values at the 50th and 51st percentiles.  
   - The difference between the 51st and 50th percentiles.

5. **Example**:

   - **Input**:  
     ```bash
     arr = [0.23456, 0.87564, 0.12345, 0.56789, 0.34567]
     ```
   - **Process**:  
     - Round off: `[0.235, 0.876, 0.123, 0.568, 0.346]`  
     - 0th Percentile: `0.123`, Minimum Value: `0.123` -> Match  
     - 100th Percentile: `0.876`, Maximum Value: `0.876` -> Match  
     - 50th Percentile: `0.346`, 51st Percentile: `0.35488`  
     - Difference: `0.008879999999999999`  

   - **Output**:  
     ```bash
     Rounded Array: [0.235, 0.876, 0.123, 0.568, 0.346]
     0th Percentile Matches Minimum: True
     100th Percentile Matches Maximum: True
     50th Percentile: 0.346, 51st Percentile: 0.35488
     Difference: 0.008879999999999999
     ```

6. **Additional Notes**:  
   - Use NumPy functions like `np.percentile` to compute percentiles.  
   - Ensure that all results are displayed with proper precision.

````python
# Solution - Question 09

arr = np.array([0.23456, 0.87564, 0.12345, 0.56789, 0.34567])

rounded_array = np.round(arr, decimals=3)
print(f'Rounded Array : {rounded_array}')

min_value = np.min(rounded_array)
zeroth_percentile = np.percentile(rounded_array, 0)
print(f'0th Percentile Mathces Minimum: {min_value == zeroth_percentile}')

max_value = np.min(rounded_array)
hundredth_percentile = np.percentile(rounded_array, 0)
print(f'100th Percentile Mathces Maximum: {hundredth_percentile == max_value}')

fifteeth_percentile = np.percentile(rounded_array, 50)
fifty_first_percentile = np.percentile(rounded_array, 51)
print(f'50th Percentile: {fifteeth_percentile}, 51st Percentile: {fifty_first_percentile}')
print(f'Difference: {fifty_first_percentile - fifteeth_percentile}')
````
